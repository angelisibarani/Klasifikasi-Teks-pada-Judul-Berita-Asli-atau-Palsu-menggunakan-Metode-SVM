```python
import pandas as pd
```

```python
data_train = pd.read_csv('D:/Materi Kuliah/Semester 9/NLP/proyek akhir/fake_or_real_news.csv')
```

```python
del data_train['text']
print (len(data_train))
data_train.head(30)
data_train.count()
```

```python
import string
for i in range(len(data_train)):
    text = str.lower(data_train['title'].iloc[i])
    data_train['title'].iloc[i]=text
print (data_train['title'][0])
data_train.head()
```

```python
#remove number
import re
pattern=r'[0-9]+'
for i in range(len(data_train)):
    data_train['title'].iloc[i] = re.sub(pattern,'', data_train['title'].iloc[i], flags=re.MULTILINE)
print (data_train['title'][0])
data_train.head()
```

```python
#remove URL
import re
pattern=r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*'
for i in range(len(data_train)):
    data_train['title'].iloc[i] = re.sub(pattern,'', data_train['title'].iloc[i], flags=re.MULTILINE)
print (data_train.title[0])
data_train.head()
```

```python
#remove RT
pattern=r'rt @\w+: '
for i in range(len(data_train)):
    data_train['title'].iloc[i] = re.sub(pattern,'', data_train['title'].iloc[i], flags=re.MULTILINE)
print (data_train['title'][0])
data_train.head()
```

```python
#remove @
pattern=r'@\w+ '
for i in range(len(data_train)):
    data_train['title'].iloc[i] = re.sub(pattern,'', data_train['title'].iloc[i], flags=re.MULTILINE)
data_train.head()
```

```python
#remove bad character
#[^A-Za-z ]
import re
pattern=r'[^A-Za-z ]'
for i in range(len(data_train)):
    data_train['title'].iloc[i] = re.sub(pattern,'', data_train['title'].iloc[i], flags=re.MULTILINE)
print (data_train['title'][0])
data_train.head()
```

```python
# remove punctuation
import string
string.punctuation
remove=string.punctuation
for i in range(len(data_train)):''
sent=data_train['title'].iloc[i]
kd=' '.join(word.strip(remove) for word in sent.split())
data_train['title'].iloc[i]=kd
print (data_train.title[0])
data_train.head()
```

```python
import re
import string
import csv
import nltk 
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
for i in range(len(data_train)):
    sent=data_train['title'].iloc[i]
    output = stemmer.stem(sent)
    data_train['title'].iloc[i]=output
print (data_train.title[0])
data_train
```

```python
# remove the duplicate titles
data_train=data_train[~data_train['title'].duplicated()]
data_train=data_train.reset_index(drop=True)
print (len(data_train))
print (data_train.title[0])
data_train.head()
```
